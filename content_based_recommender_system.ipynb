We need to now find a metric for rating the movies. So for that we need to calculate the score for each movie and sort them to recommend it for the user. For calculating the score we can use the vote_average column from our dataset. To make the scoring more accurate I will be making use of IMDB’s weighted rating formulae to calculate the score for the movies.

```python
# calculating the mean vote
c = df2['vote_average'].mean()
c
```

The result, c, will be the average of all the values in the 'vote_average' column of df2. This can be used, for instance, as the
C value in the IMDB weighted rating formula mentioned earlier, which represents the mean vote across the whole dataset.






```python
# ca;lculating the minimum votes to be listed
m = df2['vote_count'].quantile(0.9)
m
```

We are taking quartile 90 since we want movies which have got votes more than 90% to be in the list i.e. movies with higher number of votes should be included.
In the context of the IMDB weighted rating formula discussed earlier, the value of m (computed as the 90th percentile of vote counts) will be used as a threshold to determine which movies are popular or relevant enough to be considered in the top charts. This ensures that only movies with a significant number of votes (top 10% in this case) are considered, filtering out movies with very few votes regardless of their average ratings.
